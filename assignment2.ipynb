```python
df["Gender"].mode()
df["Gender"].fillna(65.8820, inplace = True)
```

```python
df["HourlyRate"].mean()
df["HourlyRate"].fillna(65.8820, inplace=True)
```

```python
df["JobInvolvement"].mean()
df["JobInvolvement"].fillna(2.7329, inplace=True)
```

```python
df["JobLevel"].mean()
df["JobLevel"].fillna(2.0611, inplace=True)
```

```python
df["JobSatisfaction"].mean()
df["JobSatisfaction"].fillna(2.7265, inplace=True)
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
```

```python
df["Department"].value_counts()
```

```python
df["Attrition"]=le.fit_transform(df["Attrition"])
```

```python
df["Department"]=le.fit_transform(df["Department"])
```

```python
df["EducationField"]=le.fit_transform(df["EducationField"])
```

```python
df["Gender"] = le.fit_transform(df["Gender"].astype(str))
```

```python
df["MaritalStatus"]=le.fit_transform(df["MaritalStatus"])
```

```python
df["OverTime"]=le.fit_transform(df["OverTime"])
```
